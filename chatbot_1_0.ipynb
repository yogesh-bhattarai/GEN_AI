Vector Embeddings

```python
# UnstructuredMarkdownLoader ko lagi loader define gareko cha
loader = UnstructuredMarkdownLoader(document_path)

# Document load garne using the loader
loaded_documents = loader.load()
```

```python
# RecursiveCharacterTextSplitter ko instance create gareko cha
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2048, chunk_overlap=128)

# Loaded documents lai split garne using the text splitter
docs = text_splitter.split_documents(loaded_documents)

# Split gareko documents ko count check garne
len(docs)
```

```python
# Paila ko split gareko document ko content print garne
print(docs[0].page_content)
```

```python
# FastEmbedEmbeddings ko instance create gareko cha
embeddings = FastEmbedEmbeddings(model_name="BAAI/bge-base-en-v1.5")
```

```python
# Qdrant ma documents ra embeddings bata embeddings store garne
qdrant = Qdrant.from_documents(
    docs,  # Split gareko documents
    embeddings,  # FastEmbedEmbeddings ko instance
    # location=":memory:",  # Optional in-memory location (commented out)
    path="./db",  # Local path jasma embeddings store huncha
    collection_name="document_embeddings",  # Collection ko naam
)
```

```python
# Time measurement ko lagi cell magic command
%%time

# Query define gareko cha
query = "Can a local resident who is permanently settled get citizenship?"

# Similarity search garne using Qdrant with the given query
similar_docs = qdrant.similarity_search_with_score(query)
```

```python
# Similar documents ra uniharu ko score print garne
for doc, score in similar_docs:
    # Document ko text ko pahilo 256 characters print garne
    print(f"text: {doc.page_content[:256]}\n")

    # Document ko score print garne
    print(f"score: {score}")

    # Divider line print garne for better readability
    print("-" * 80)
    print()  # Extra newline for spacing
```

```python
# Time measurement ko lagi cell magic command
%%time

# Qdrant ko retriever banaune with search parameters
retriever = qdrant.as_retriever(search_kwargs={"k": 5})  # Top 5 similar documents retrieve garne

# Query ko adhar ma documents retrieve garne
retrieved_docs = retriever.invoke(query)
```

```python
# Retrieved documents ko details print garne
for doc in retrieved_docs:
    # Document ko ID print garne
    print(f"id: {doc.metadata['_id']}\n")

    # Document ko text ko pahilo 256 characters print garne
    print(f"text: {doc.page_content[:256]}\n")

    # Divider line print garne for better readability
    print("-" * 80)
    print()  # Extra newline for spacing
```

Reranking

```python
# Flashrank reranking ko compressor create gareko cha
compressor = FlashrankRerank(model="ms-marco-MiniLM-L-12-v2")

# ContextualCompressionRetriever ko instance create gareko cha
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor,  # Compressor ko base instance
    base_retriever=retriever  # Base retriever ko instance
)
```

```python
# Time measurement ko lagi cell magic command
%%time

# Query ko adhar ma reranked documents retrieve garne
reranked_docs = compression_retriever.invoke(query)

# Reranked documents ko count check garne
len(reranked_docs)
```

```python
# Reranked documents ko details print garne
for doc in reranked_docs:
    # Document ko ID print garne
    print(f"id: {doc.metadata['_id']}\n")

    # Document ko text ko pahilo 256 characters print garne
    print(f"text: {doc.page_content[:256]}\n")

    # Document ko relevance score print garne
    print(f"score: {doc.metadata['relevance_score']}")

    # Divider line print garne for better readability
    print("-" * 80)
    print()  # Extra newline for spacing
```

```python
# ChatGroq ko instance create gareko cha
llm = ChatGroq(temperature=0, model_name="llama3-70b-8192")
```

```python
# User ko prashna ko uttar dinu parne template define gareko cha
prompt_template = """
Use the following pieces of information to answer the user's question.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {context}
Question: {question}

Answer the question and provide additional helpful information,
based on the pieces of information, if applicable. Be succinct.

Responses should be properly formatted to be easily read.
"""

# PromptTemplate ko instance create gareko cha
prompt = PromptTemplate(
    template=prompt_template,  # Template ko string
    input_variables=["context", "question"]  # Input variables specify garne
)
```

```python
# RetrievalQA ko instance create gareko cha using specified LLM and retriever
qa = RetrievalQA.from_chain_type(
    llm=llm,  # ChatGroq ko instance
    chain_type="stuff",  # Chain type specify gareko cha
    retriever=compression_retriever,  # Contextual compression retriever
    return_source_documents=True,  # Source documents return garne
    chain_type_kwargs={"prompt": prompt, "verbose": True},  # Additional parameters
)
```

```python
# Time measurement ko lagi cell magic command
%%time

# User ko prashna ko uttar paune process run gareko cha
response = qa.invoke("To give Nepali citizenship certificate to citizens based on what documents?")
```

```python
print_response(response)
```
